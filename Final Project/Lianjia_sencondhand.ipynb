```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url="https://hangzhou.anjuke.com/community/?from=esf_list_navigation"
headers = {'User-Agent':'User-Agent:Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36'}
r = requests.get(url,headers=headers).text
print(r)
# r.text
mypage=BeautifulSoup(r.text)
```

```python
urls=[]
houseprice=[]
housename=[]
location=[]
time_of_completion=[]

for i in range (0,5):
    url='https://hangzhou.anjuke.com/community/p'+str(i+1)+'/'
    urls.append(url)
    
for url in urls:
    page_name =requests.get(url,headers=headers).text
    mypage=BeautifulSoup(page_name,'html.parser')
    
    names=mypage.find_all('h3')
    name=names[0]
    for name in names:
        housename.append(name.text)

    addresses=mypage.find_all('address')
    address=addresses[0]
    for address in addresses:
        location.append(address.text)
    
    prices=mypage.find_all('strong')
    price=prices[0]
    for price in prices:
        houseprice.append(price.text)
    
    times=mypage.find_all('p',attrs={'class':'date'})
    time=times[0]
    for time in times:
        time_of_completion.append(time)
```

```python
with open("Hangzhou_secondhand_house.csv",'w',newline='') as f:
    spamwriter = csv.writer(f)
    header=['housename','location','houseprice','time_of_completion']
    spamwriter.writerow(header)
    spamwriter.writerows(zip(housename,location,houseprice,time_of_completion))
    
    
```
